#Filtering Data

```python
df["mark"]>90
# Only returns the boolean value
```

```python
# df[condition]
# Mark greater than 90
df[df["mark"]>90]
# this code will return only true
```

```python
# Return female students only
df[df["gender"]=="female"]
```

```python
df.head()
```

```python
df[df["class"]=="Four"]
# Returns all the reults for students in class 4
```

```python
df[df["class"]=="Four"][['name', 'class', 'gender']]
 #Using both filtering *df[df["class"]=="Four"]* and slicing [['name', 'class', 'gender']]
 # returns only name, class and gender for students in class 4.
```

```python
df.loc[df["class"]=="Four", ["name", "class", "gender"]]
# same as above but using the loc function.
```

```python
df.iloc[df['class']=="Four",[1, 2, 4]]
# iloc doesnt support filtering.
```

* when using multiple columns to filter, enclose each condtion by () and seperate them by &

```python
df[(df["class"] =="Four") & (df['gender'] =="male")]
```

```python
df[(df["mark"]<70) & (df["gender"]=="female")][['name', 'class']]
# returns the name and class of females with marks less than 70
```

```python
df.loc[(df["mark"]<70) & (df["gender"]=="female"),["name", "class"]]
# Same thing as above but using loc operator.
```

#Sorting results

* There are always the natural order of the rows. In this instance student info is sorted by id.

```python
df.head ()
```

```python
df.sort_values(by="name")
# sorts the names in alphabetical order
```

```python
df.sort_values(by="mark").head()
# if sorted by numeric value it goes in ascending order. because of .head()
# will return the first 5 results in numerical order ascending.
```

```python
df.sort_values(by="name", ascending=False).head(10)
# descending order
# ascending false will reverse the order. its in descending order
```

```python
df.sort_values(by="mark", ascending=False).head(10)
# ascending order
```

```python
# applyting sort after slicing
df[df["class"]=="Six"].sort_values(by="mark", ascending=False).head(10)
```

```python
df.loc[df["class"]=="Six", :].sort_values(by="mark")
# same as above using df.loc function
```

```python
df[(df["gender"] == "female") & (df["mark"] > df["mark"].mean())].sort_values(by="mark", ascending=False)
# returns female students that have a mark higher than the mean average.
```
